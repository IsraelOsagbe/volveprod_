```python
#new and easier implementation
import plotly.express as px
#old implementations
import plotly.graph_objects as go
```

```python
import pandas as pd

filename = 'volve.xlsx'

df_daily = pd.read_excel(filename, sheet_name = "Daily Production Data")
df_monthly = pd.read_excel(filename, sheet_name = "Monthly Production Data")
```

```python
#return the column names in the first sheet

cols= df_daily.columns
cols
```

```python
df_daily['Wellbore name'].unique()
```

```python
#creating simple plots
px.line(df_daily, x = 'DATEPRD', y= 'BORE_OIL_VOL') #this includes all seven wells
```

```python
#we need to categorise the plots based on the unique wells
px.line(df_daily, x = 'DATEPRD', y= 'BORE_OIL_VOL', color='Wellbore name') 
```

```python
#to plot a signle well, you need to create a filter for that particular well

df_well1 = df_daily[df_daily['Wellbore name'] == '15/9-F-1 C']
px.line(df_well1, x = 'DATEPRD', y= ['BORE_OIL_VOL', 'BORE_WAT_VOL']) 
```

```python
import matplotlib.pyplot as plt
px.line( df_daily, x = 'DATEPRD', y = ['BORE_OIL_VOL', 'BORE_WAT_VOL'], facet_col= 'Wellbore name', facet_col_wrap = 4,
        width = 1500, height= 820 )
# plt.savefig('Daily Production Data by Wells.pdf')
```
